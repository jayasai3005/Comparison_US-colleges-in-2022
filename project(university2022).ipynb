```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("2022 US college Rankings.csv")
```

```python
sorted_Tution_df=df.sort_values(by='Tuition')
```

```python
sorted_Rank_df=df.sort_values(by='Adjusted Rank')
```

```python
#Top 10 college by Rank
sorted_Rank_df.head(10)
```

```python
ax=sns.countplot(x='Adjusted Rank',data=sorted_Rank_df)
for bars in ax.containers:
    ax.bar_label(bars)
    ax.set_xlim(0,10)
    ax.set_ylim(0,6)
    
```

```python
ax=sns.lineplot(x='Adjusted Rank',y='Tuition',data=sorted_Rank_df)
for bars in ax.containers:
    ax.bar_label(bars)
```
